## 1. Goldman Sachs - Greece Currency Swap

**1.1 Fair Swap Rate**

```python
import pandas as pd
import numpy as np

df_zcb = pd.read_excel("Assignments/Assignment 3/Greece_GS_table1.xls", header=None)
df_zcb = df_zcb.iloc[2:].copy()
df_zcb.columns = ['NaN', 'Maturity', 'ZEU', 'ZUS']
df_zcb = df_zcb[['Maturity', 'ZEU', 'ZUS']].reset_index(drop=True)

S0 = 0.8475
pv_usd_leg = 0
for idx, row in df_zcb.iterrows():
    t = row['Maturity']
    if t == 0.0: continue
    cf = 1.5
    if t == 10.0: cf += 50.0
    pv_usd_leg += cf * row['ZUS']

sum_zeu = df_zcb[df_zcb['Maturity'] > 0]['ZEU'].sum()
zeu_T = df_zcb.iloc[-1]['ZEU']

c_eur_fair = (pv_usd_leg / S0 - 59.0 * zeu_T) / (29.5 * sum_zeu)
print(f"Fair EUR Rate: {c_eur_fair:.4%}")
```

**1.2 Goldman Deal Value**

```python
S_off = 0.8148
Notional_EUR_GS = 50.0 / S_off
c_gs = 0.07

pv_eur_out_gs = (c_gs / 2 * Notional_EUR_GS) * sum_zeu + Notional_EUR_GS * zeu_T
val_gs_deal = (pv_usd_leg - 50.0) + S0 * (Notional_EUR_GS - pv_eur_out_gs)
print(f"Value of Deal to Greece: {val_gs_deal:.3f} Billion USD")
```

## 2. Hedging with Options

**2.3 Implied Interest Rate**

```python
C = 2.541
P = 11.366
K = 105
S = 95
T = 0.25

# C - P = S - K * D
D = (S - (C - P)) / K
r = -np.log(D) / T
print(f"Implied Rate: {r:.4%}")
```

## 1. Barings / Leeson position on Nikkei options

**Analysis of Strategy**
Leeson sold 35,500 Calls and 35,500 Puts (Short Straddle).

```python
K = 19750
c = 990
p = 980
Total_Premium = (c + p) * 10000 * 35500
S_Final = 17473
Payoff_Unit = abs(S_Final - K)
Loss_Total = (c + p - Payoff_Unit) * 10000 * 35500

print(f"Total Premium Received: {Total_Premium:,.0f} JPY")
print(f"Final P&L: {Loss_Total:,.0f} JPY")
```

## 2. Binomial Trees - Vanda Pharmaceutical

```python
import numpy as np

S1_u = 21.0
S1_d = 10.0
q_real = 0.7
beta = 2.0
r_cc = 0.05
E_RP = 0.0644

r_annual = np.exp(r_cc) - 1
E_R_annual = r_annual + beta * E_RP
E_S1_analyst = q_real * S1_u + (1 - q_real) * S1_d
S0 = E_S1_analyst / (1 + E_R_annual)

print(f"S0: {S0:.4f}")

# Option Value
K = S0
Cu = max(S1_u - K, 0)
Cd = max(S1_d - K, 0)
p_rn = (S0 * np.exp(r_cc) - S1_d) / (S1_u - S1_d)
V_opt = np.exp(-r_cc) * (p_rn * Cu + (1 - p_rn) * Cd)

print(f"Option Value: {V_opt:.4f}")
```

## 2. Binomial Trees

**Valuation**

```python
import numpy as np

S0 = 100
r_cc = 0.02
dt = 0.5

# Tree Params
u1 = 1.20; d1 = 0.90
u2 = 1.15; d2 = 0.95

# RN Probs
p1 = (np.exp(r_cc*dt) - d1)/(u1 - d1)
p2 = (np.exp(r_cc*dt) - d2)/(u2 - d2)

# Terminal Values
# Nodes: UU, UD, DU, DD
S_UU = S0 * u1 * u2
S_UD = S0 * u1 * d2
S_DU = S0 * d1 * u2
S_DD = S0 * d1 * d2

# Payoffs (Call 120, Put 100)
def payoff(S):
    c = max(S - 120, 0)
    p = max(100 - S, 0)
    return c + p

V_UU = payoff(S_UU)
V_UD = payoff(S_UD)
V_DU = payoff(S_DU)
V_DD = payoff(S_DD)

# Exp Value
E_V = p1*p2*V_UU + p1*(1-p2)*V_UD + (1-p1)*p2*V_DU + (1-p1)*(1-p2)*V_DD
Price = E_V * np.exp(-r_cc * 1.0)

print(f"Option Package Price: {Price:.4f}")
```

## 4. Swap

**Calculations**

```python
swap_rate = 1.209
fwd_05 = 1.206
r_05 = 0.05
r_10 = 0.06

df_05 = np.exp(-r_05 * 0.5)
df_10 = np.exp(-r_10 * 1.0)

F_1 = (swap_rate * (df_05 + df_10) - fwd_05 * df_05) / df_10
print(f"1Y Forward Rate: {F_1:.4f}")
```
